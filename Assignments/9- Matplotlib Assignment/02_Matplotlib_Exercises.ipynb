## Exercise 1

** Follow along with these steps: **
* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 100, 100)
y = 2 * x

fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, 'b')

ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_title("X vs Y Plot")

plt.show()
```

** Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

** Now plot (x,y) on both axes. And call your figure object to show it.**

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax1 = plt.axes([0, 0, 1, 1])

ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

x = np.arange(0, 100)
y = 2 * x

ax1.plot(x, y, color='blue', linewidth=3)
ax1.set_title('Main Plot (ax1)')
ax1.set_xlabel('X for ax1')
ax1.set_ylabel('Y for ax1')

ax2.plot(x, y, color='red')
ax2.set_title('Inset Plot (ax2)')
ax2.set_xlabel('Y for ax2')
ax2.set_ylabel('X for ax2')

plt.show()
```

## Exercise 3

** Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8, 6), dpi=100)

ax1 = fig.add_axes([0, 0, 1, 1])

ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

ax1.set_title('Main Plot (ax1)', fontsize=14)
ax1.set_xlabel('X for ax1', fontsize=12)
ax1.set_ylabel('Y for ax1', fontsize=12)


ax1.set_xlim(0, 1)
ax1.set_ylim(0, 1)

ax2.set_title('Inset Plot (ax2)', fontsize=12)
ax2.set_xlabel('Y for ax2', fontsize=10) 
ax2.set_ylabel('X for ax2', fontsize=10) 

ax2.set_xlim(0, 1)
ax2.set_ylim(0, 1)

plt.show()

```

** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 100, 100)
z = x**2

fig = plt.figure(figsize=(8, 6), dpi=100)

ax1 = fig.add_axes([0, 0, 1, 1])

ax2 = fig.add_axes([0.25, 0.5, 0.4, 0.35])

ax1.plot(x, z, color='blue')

ax1.set_xlabel('X', fontsize=12)
ax1.set_ylabel('Z', fontsize=12)

x_zoom = np.linspace(20, 22, 100)
y_zoom = x_zoom**2
ax2.plot(x_zoom, y_zoom, color='blue')

ax2.set_title('zoom', fontsize=12)
ax2.set_xlabel('X', fontsize=10)
ax2.set_ylabel('Y', fontsize=10)

ax2.set_xlim(20, 22)
ax2.set_ylim(30, 50)

plt.show()

```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 100, 100)
y = 2 * x
z = x ** 2

fig, ax = plt.subplots(nrows=1, ncols=2)

ax[0].plot(x, y, color='blue', linewidth=3, linestyle='--')
ax[0].set_title('X vs Y Plot (Dashed)')
ax[0].set_xlabel('X')
ax[0].set_ylabel('Y')

ax[1].plot(x, z, color='red', linewidth=1, linestyle=':')
ax[1].set_title('X vs Z Plot (Dotted)')
ax[1].set_xlabel('X')
ax[1].set_ylabel('Z')

plt.tight_layout()

plt.show()
```

** See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 100, 100)
y = 2 * x
z = x ** 2

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

ax[0].plot(x, y, color='blue', linewidth=3, linestyle='--')
ax[0].set_title('X vs Y Plot (Dashed)')
ax[0].set_xlabel('X')
ax[0].set_ylabel('Y')

ax[1].plot(x, z, color='red', linewidth=1, linestyle=':')
ax[1].set_title('X vs Z Plot (Dotted)')
ax[1].set_xlabel('X')
ax[1].set_ylabel('Z')

plt.tight_layout()

plt.show()
```
