```python
#EX 1: 


# 1. Count the number of vowels in a given string.
def count_vowels(s: str) -> int:
    vowels = set('aeiouAEIOU')
    return sum(1 for char in s if char in vowels)

txt1 = input("Enter a Text to count the vowels: ")
print("1. the vowels count is:" , count_vowels(txt1))

# 2. Return the same string but with every second letter capitalized.
def capitalize_every_second(s: str) -> str:
    result = []
    for i, ch in enumerate(s):
        if i % 2 == 1:
            result.append(ch.upper())
        else:
            result.append(ch)
    return ''.join(result)

txt2 = input("Enter word to capitalize every second letter: ")
print("2. the capitalized string is:" , capitalize_every_second(txt2))

# 3. Return a new list containing only the even numbers from a list.
def filter_even(numbers: list[int]) -> list[int]:
    return [n for n in numbers if n % 2 == 0]

txt3 = input("Enter numbers to filter the even numbers: ")

numbers = [int(x) for x in txt3.split()]

print("3. the evens in the numbers is:" , filter_even(numbers))

# 4. Return a new list with duplicate values removed (preserving order).
def remove_duplicates(lst: list) -> list:
    seen = set()
    result = []
    for item in lst:
        if item not in seen:
            seen.add(item)
            result.append(item)
    return result

txt4 = input("Enter numbers to remove the duplicate values: ")

numbers1 = [int(x) for x in txt4.split()]

print("4. the set with no duplicates is:" , remove_duplicates(numbers1))

# 5. Return the intersection of two sets as a new set.
def set_intersection(a: set, b: set) -> set:
    return a & b

txt51 = input("Enter the first groub of numbers for intersection: ")
numbers2 = map(int, txt51.split())
nSet = set(numbers2)

txt52 = input("Enter the second groub of numbers for intersection: ")
numbers3 = map(int, txt52.split())
nSet1 = set(numbers3)

print("5. the intersection is" , set_intersection(nSet, nSet1))

# 6. Return the union of two sets as a new set.
def set_union(a: set, b: set) -> set:
    return a | b

txt61 = input("Enter the first groub of numbers for union: ")
numbers4 = map(int, txt61.split())
nSet2 = set(numbers4)

txt62 = input("Enter the second groub of numbers for union: ")
numbers5 = map(int, txt62.split())
nSet3 = set(numbers5)

print("6. the union is:" , set_union(nSet2, nSet3))

# 7. Sort a list of tuples based on the second element of each tuple.
def sort_by_second(tuples: list[tuple]) -> list[tuple]:
    return sorted(tuples, key=lambda x: x[1])

txt7 = input("Enter tuples separated by commas (e.g. (1,3),(2,1),(3,2)): ")

tuples_list = [tuple(map(int, t.strip(" ()").split(','))) for t in txt7.split('),')]

print("7. Sorted by second element:", sort_by_second(tuples_list))


# 8. Return a reversed version of a tuple.
def reverse_tuple(t: tuple) -> tuple:
    return t[::-1]

txt8 = input("Enter tuple elements separated by commas (e.g. 1,2,3): ")

input_tuple = tuple(map(int, txt8.split(',')))

print("8. Reversed tuple:", reverse_tuple(input_tuple))


# 9. Return the keys and values of a dictionary as separate lists.
def dict_to_lists(d: dict) -> tuple[list, list]:
    keys = list(d.keys())
    values = list(d.values())
    return keys, values

txt9 = input("Enter dictionary items as key:value pairs separated by commas (e.g. a:1,b:2): ")

d = {}
for item in txt9.split(','):
    key1, value1 = item.split(':')
    key1 = key1.strip()
    value1 = value1.strip()

    if value1.isdigit():
        value1 = int(value1)
    d[key1] = value1

print("9. Keys and values:", dict_to_lists(d))


# 10. Return d[key] if key exists, else return default.
def get_with_default(d: dict, key, default=0):
    return d.get(key, default)

txt10 = input("Enter dictionary items as key:value pairs separated by commas (e.g. x:10,y:20): ")
d = {}
for item in txt10.split(','):
    key2, value2 = item.split(':')
    key2 = key2.strip()
    value2 = value2.strip()

    if value2.isdigit():
        value2 = int(value2)
    d[key2] = value2

key_input = input("Enter the key to look up: ").strip()

print("10. Result:", get_with_default(d, key_input))



```
